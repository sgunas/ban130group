```python
df = pd.read_sas('np_summary.sas7bdat')
df['Type'] = df['Type'].str.decode('utf-8') #Type column was stored in "bytes" thefore I converted into str

#Add a new column ParkType using conditional logic:
df['ParkType'] = 'Other'

df.loc[df['Type'] == 'NM', 'ParkType'] = 'Monument'
df.loc[df['Type'] == 'NP', 'ParkType'] = 'Park'
df.loc[df['Type'].isin(['NPRE', 'PRE', 'PRESERVE']), 'ParkType'] = 'Preserve'
df.loc[df['Type'] == 'NS', 'ParkType'] = 'Seashore'
df.loc[df['Type'].isin(['RVR', 'RIVERWAYS']), 'ParkType'] = 'River'


# Generate frequency report
freq_report = df["ParkType"].value_counts()

print(freq_report)

# Create bar chart
freq_report.plot(kind="bar")

plt.title("Frequency of Park Types")
plt.xlabel("Park Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# This section reloads the np_summary dataset and creates a new ParkType column. 
# Conditional logic is used to convert park type codes into readable labels such as Monument, Park, Preserve, Seashore, and River. 
# A frequency report is created to count how many parks fall into each ParkType category. 
# A bar chart is generated to visually display the ParkType frequency results.
```

```python
df = pd.read_sas("customer.sas7bdat")

query1 = df[(df["BankID"].isna()) & (df["CreditScore"] > 700)]
# Sort by Income descending
query1 = query1.sort_values(by="Income", ascending=False)

# Format Income in dollar format
query1["Income"] = query1["Income"].apply(lambda x: f"${x:,.2f}")

#nprint income column to check 
print("Query 1 Results:")
print(query1[['Income']])

# Employed column was stored in bytes thefore I converted into str in order to be opertable
df["Employed"] = df["Employed"].str.decode('utf-8').str.strip()

# Convert SAS date format (days since 1960-01-01) to datetime
df['DOB'] = pd.to_datetime(df['DOB'], unit='d', origin='1960-01-01')

# Filter rows where DOB is earlier than 1940-12-31 and Employed is Y
query2 = df[(df['DOB'] < '1940-12-31') & (df['Employed'] == 'Y')]

# Sort by DOB descending
query2 = query2.sort_values(by='DOB', ascending=False)

# print DOB and Employed Column to check 
print("Query 2 Results:")
print(query2[["DOB", "Employed"]])

```

```python
# This section loads the customer SAS dataset for customer-based filtering and reporting. 
# The first query filters customers where BankID is missing and CreditScore is greater than 700. 
# The results are sorted by Income from highest to lowest, and Income is formatted in dollar format. 
# The second query filters customers with a DOB before 1940-12-31 and Employed equal to Y. 
# These results are sorted by DOB in descending order to show the most recent dates first.
```

```python
# Load the customer dataset
customer = pd.read_sas("customer.sas7bdat")
```

```python
# This section uses the customer dataset to create new business-friendly categories.
```

```python
# Create CreditCategory column
conditions = [
    customer["CreditScore"] >= 750,
    customer["CreditScore"].between(700, 749),
    customer["CreditScore"].between(650, 699),
    customer["CreditScore"].between(550, 649),
    customer["CreditScore"] < 550
]

choices = [
    "Excellent",
    "Good",
    "Fair",
    "Poor",
    "Bad"
]

customer["CreditCategory"] = np.select(
    conditions,
    choices,
    default="Unknown"
)
```

```python
# A CreditCategory column is created by grouping customers based on their CreditScore.
# Customers are classified as Excellent, Good, Fair, Poor, or Bad based on the required score ranges.
```

```python
# Create MaritalStatus column using the existing Married column
customer["MaritalStatus"] = customer["Married"].map({
    b"M": "Married",
    b"D": "Divorced",
    b"S": "Single",
    b"W": "Widowed"
}).fillna("Unknown")
```

```python
# A MaritalStatus column is created by converting marital status codes into readable labels. 
```

```python
# Display results
print(customer[["CreditScore", "CreditCategory", "Married", "MaritalStatus"]].head(10))
```

```python
# A MaritalStatus column is created by converting marital status codes into readable labels. 
# Any marital status code outside M, D, S, or W is labelled as Unknown.
```
